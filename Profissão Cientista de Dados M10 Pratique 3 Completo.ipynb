````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
arquivo = pd.read_csv('previsao_de_renda.csv', encoding = 'latin1')
df = pd.DataFrame(arquivo)
df.dtypes
````

````python
df['data_ref'] = pd.to_datetime(df['data_ref'])
df_qual = df.select_dtypes(include=['object','bool']).columns.tolist()
df_qual
````

````python
tab_sexo = pd.crosstab(df['data_ref'], df['sexo'])
tab_sexo.div(tab_sexo.sum(axis = 1), axis = 0)

ax = tab_sexo.plot.bar(stacked = True)
tick_labs = df['data_ref'].map(lambda ts: ts.strftime('%m-%Y')).unique()
ticks = ax.set_xticks(list(range(df['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=90)

plt.title('Proporção variável sexo ao Longo do Tempo')
plt.xlabel('Data de Refênceria')
plt.ylabel('Sexo')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.figure(figsize=(400, 400))
plt.tight_layout()
plt.show()
````

2) Para cada variável qualitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?

````python
for coluna in df_qual:
    grouped = df.groupby(['data_ref', coluna])['renda'].mean().reset_index()
    
    plt.figure(figsize=(14, 6))
    sns.lineplot(data=grouped, x='data_ref', y='renda', hue=coluna, marker='o')
    plt.title(f'Média de Renda por {coluna} ao Longo do Tempo')
    plt.xlabel('Data de Refênceria')
    plt.ylabel('Renda Média (R$)')
    plt.xticks(rotation=45)
    plt.legend(title=coluna, bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()
````

Penso que a alteração mais importante foi a forma de visualizar o comportamento da renda nas variáveis qualitativas. Dessa forma é possível observar informações como os picos de maior renda para clientes com pós-graduação e a queda de renda em bolsistas.
Creio que é possível usar essas variáveis em uma modelagem, pois elas possuem comportamentos individuais em relação a renda, podendo nos trazer uma indicação se o cliente será um bom ou mau pagador.
