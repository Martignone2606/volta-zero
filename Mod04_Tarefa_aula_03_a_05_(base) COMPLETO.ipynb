## Exercício 1: Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
import pandas as pd
import numpy as np

np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])
````

````python
#Como não foi solicitado para criar um novo dataframe, irei fazer das 2 formas, 
#sem criar um dataframe para a soma e criando um dataframe para a soma
#Primeiro sem criar um dataframe:
````

````python
# 1) Seu código aqui
df_mat['% mat'] = np.round(((df_mat['Qt_acertos']*100)/24),2)
df_por['% por'] = np.round(((df_por['Qt_acertos']*100)/18),2)
df_geo['% geo'] = np.round(((df_geo['Qt_acertos']*100)/8),2)
df_ing['% ing'] = np.round(((df_ing['Qt_acertos']*100)/8),2)
df_his['% his'] = np.round(((df_his['Qt_acertos']*100)/8),2)
df_fis['% fis'] = np.round(((df_fis['Qt_acertos']*100)/12),2)
df_qui['% qui'] = np.round(((df_qui['Qt_acertos']*100)/12),2)
df_fis
````

````python
# 2) Seu código aqui
df_mat['Qt_acertos'] + df_por['Qt_acertos'] + df_geo['Qt_acertos']
+ df_ing['Qt_acertos'] + df_his['Qt_acertos'] + df_fis['Qt_acertos'] + df_qui['Qt_acertos']
````

````python
# 3) Seu código aqui
np.round(((df_mat['Qt_acertos'] + df_por['Qt_acertos'] + df_geo['Qt_acertos']
+ df_ing['Qt_acertos'] + df_his['Qt_acertos'] + df_fis['Qt_acertos'] + df_qui['Qt_acertos'])*100/90),2)
````

````python
# 4) Seu código aqui
aprovados = np.round((df_mat['Qt_acertos'] + df_por['Qt_acertos'] + df_geo['Qt_acertos']
+ df_ing['Qt_acertos'] + df_his['Qt_acertos'] + df_fis['Qt_acertos'] + df_qui['Qt_acertos']))
aprovados

aprovado = 0
for lin in aprovados:
    if lin >= 45:
        aprovado += 1

print ('Total de aprovados para a segunda fase:',aprovado)
````

````python
#Criando um dataframe
````

````python
df_materias = pd.DataFrame({'Total_acertos': df_mat['Qt_acertos'] + df_por['Qt_acertos'] + df_geo['Qt_acertos']
                         + df_ing['Qt_acertos'] + df_his['Qt_acertos'] + df_fis['Qt_acertos'] + df_qui['Qt_acertos']})
df_materias
````

````python
df_materias['% Total'] = np.round(((df_materias['Total_acertos']*100)/90),2)
df_materias
````

````python
df_materias['Aprovado'] = 0
df_materias.loc[df_materias['Total_acertos'] >= 45, 'Aprovado'] = 1
print('Total de aprovados para a segunda fase:',df_materias['Aprovado'].sum())
````

````python
df_materias.loc[df_materias['Total_acertos'] >= 45]
````

## 2) Vestibular II

Ainda sobre o mesmo banco de dados:

1. Neste vestibular, quem 'zera' em matemática, física ou química está desqualificado. Monte um novo *DataFrame* com os alunos desqualificados por este critério.
2. Quantos são esses alunos?
3. Qual a média desses alunos em história e geografia?
4. Monte um *DataFrame* com os alunos que passaram para a segunda fase. Repare que estes alunos não podem ter sido desqualificados.

````python
# seu código aqui
df_desqualificados = pd.DataFrame()
df_desqualificados['Matematica'] = df_mat['Qt_acertos']
df_desqualificados['Fisica'] = df_fis['Qt_acertos']
df_desqualificados['Quimica'] = df_qui['Qt_acertos']

df_desqualificados['Desqualificado'] = 0
df_desqualificados.loc[df_desqualificados['Matematica'] == 0, 'Desqualificado'] = 1
df_desqualificados.loc[df_desqualificados['Fisica'] == 0, 'Desqualificado'] = 1
df_desqualificados.loc[df_desqualificados['Quimica'] == 0, 'Desqualificado'] = 1
df_desqualificados.loc[df_desqualificados['Desqualificado'] == 1]
````

````python
#Quantidade de alunos desqualificados
print('Total de desqualificados:',df_desqualificados['Desqualificado'].sum())
````

````python
#Não ficou claro se quer a média de história e geografia individual ou das 2 juntas, então farei os 2
df_desqualificados['Historia'] = df_his['Qt_acertos']
df_desqualificados['Geografia'] = df_geo['Qt_acertos']
df_desqualificados['% geo'] = np.round(((df_desqualificados['Geografia']*100)/8),2)
df_desqualificados['% his'] = np.round(((df_desqualificados['Historia']*100)/8),2)
df_desqualificados['% geo e his'] = np.round((((df_desqualificados['Geografia'] + df_desqualificados['Historia'])*100)/16),2)
df_desqualificados.loc[df_desqualificados['Desqualificado'] == 1]
````

````python
#Alunos que passsaram para a segunda fase
df_aprovados = pd.DataFrame()
df_aprovados['Matematica'] = df_mat['Qt_acertos']
df_aprovados['Fisica'] = df_fis['Qt_acertos']
df_aprovados['Quimica'] = df_qui['Qt_acertos']
df_aprovados['Historia'] = df_his['Qt_acertos']
df_aprovados['Geografia'] = df_geo['Qt_acertos']
df_aprovados['Portugues'] = df_por['Qt_acertos']
df_aprovados['Ingles'] = df_ing['Qt_acertos']
df_aprovados['Total_Acertos'] = (df_mat['Qt_acertos'] + df_fis['Qt_acertos'] + df_qui['Qt_acertos'] 
                                 + df_his['Qt_acertos'] + df_geo['Qt_acertos'] + df_por['Qt_acertos'] + df_ing['Qt_acertos'])

df_aprovados['Aprovado'] = 0
df_aprovados.loc[df_aprovados['Total_Acertos'] >= 45, 'Aprovado'] = 1
df_aprovados.loc[df_aprovados['Matematica'] == 0, 'Aprovado'] = 0
df_aprovados.loc[df_aprovados['Fisica'] == 0, 'Aprovado'] = 0
df_aprovados.loc[df_aprovados['Quimica'] == 0, 'Aprovado'] = 0
df_aprovados.loc[df_aprovados['Aprovado'] == 1]
````
