1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
````

````python
arquivo = pd.read_csv('previsao_de_renda.csv', encoding = 'latin1')
df = pd.DataFrame(arquivo)
df.head()
````

````python
ax = sns.pointplot(x='posse_de_imovel', y='renda',
                  data = df, dodge=True,errorbar=('ci', 95))
````

````python
ax = sns.pointplot(x='posse_de_veiculo', y='renda',
                  data = df, dodge=True,errorbar=('ci', 95))
````

````python
ax = sns.pointplot(x='posse_de_imovel', y='renda', hue='posse_de_veiculo',
                  data = df, dodge=True,errorbar=('ci', 95))
plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30), ncol =3)
````

````python
ax = sns.pointplot(x='posse_de_veiculo', y='renda', hue='posse_de_imovel',
                  data = df, dodge=True,errorbar=('ci', 95))
plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30), ncol =3)
````

Ambas as variáveis demonstram que há uma diferença significante de renda quando a pessoa possui posse de veículo ou de imóvel e essa diferença também fica mais clara quando possuem ambos.
