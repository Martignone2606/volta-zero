2. Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com *status* em 'Arquivado por decurso de prazo' estão registradas?

4. Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'

````python
#1) Seu código aqui
def consulta_api(uf , interessado):
  uf = input('Escreva a sigla do estado que deseja consultar: ')
  uf = uf.upper()
  interessado = input('Escolha entre o tipo interessado entre Estado ou Município: ')
  interessado = interessado.lower().capitalize()
  url_uf_interessado = 'http://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf='+uf+'&tipo_interessado='+interessado
  r_url = requests.get(url_uf_interessado)
  data_json_consulta = r_url.json()
  df_consulta = pd.DataFrame(data_json_consulta['items'])
  return df_consulta
````
